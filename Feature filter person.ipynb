```python
import numpy as np
import pandas as pd
import os
```

```python
select_files=['VEHICLE.csv','PERSON.csv','DRIMPAIR.csv',]
#fixed_path='C:\\Users\\Duong Hung\\Documents\\GitHub\\Vehicle-Crash-Data-Visualization'
```

# Vehicle.csv

```python
select_features=['STATE','ST_CASE','VEH_NO','L_TYPE','L_STATUS','DR_HGT','DR_WGT',
                'PREV_ACC', 'PREV_DWI', 'PREV_SPD', 'PREV_OTH' ]
```

```python
df_final=pd.DataFrame()
for year in range(2010,2019):
    path='.\Datasets\\'+str(year)+'\\'+select_files[0]
    print(path)
    df=pd.read_csv(path,encoding = "ISO-8859-1", low_memory=False)
    df=df[select_features]
    df['YEAR']=year
    df_final=pd.concat([df_final,df]).astype('Int32')
df_final
```

# Person.csv

```python
select_features=['STATE','ST_CASE','VEH_NO','PER_TYP','AGE','SEX','RACE',
                'INJ_SEV','DRINKING','DRUGS','WORK_INJ']
```

```python
df_final_person=pd.DataFrame()
for year in range(2010,2019):
    path='.\Datasets\\'+str(year)+'\\'+select_files[1]
    print(path)
    df=pd.read_csv(path,encoding = "ISO-8859-1", low_memory=False)
    df=df[select_features]
    df['YEAR']=year
    df_final_person=pd.concat([df_final_person,df]).astype('Int32')
df_final_person
```

```python
df_final_person=df_final_person[df_final_person.PER_TYP==1]
df_final_person.drop(columns='PER_TYP',inplace=True)
```

```python
df_final_person
```

```python
df_final=pd.merge(df_final,df_final_person,how='right',on=['STATE','ST_CASE','VEH_NO','YEAR']).astype('Int32')
df_final
```

# DRIMPAIR.csv

```python
df_final_impair=pd.DataFrame()
for year in range(2010,2019):
    path='.\Datasets\\'+str(year)+'\\'+select_files[2]
    print(path)
    df=pd.read_csv(path,encoding = "ISO-8859-1", low_memory=False)
    df['YEAR']=year
    df_final_impair=pd.concat([df_final_impair,df]).astype('Int32')
df_final_impair
```

```python
df_final=pd.merge(df_final,df_final_impair,how='left',on=['STATE','ST_CASE','VEH_NO','YEAR']).astype('Int32')
df_final
```

# Violatn.csv

```python
df_final_vio=pd.DataFrame()
for year in range(2010,2019):
    path='.\Datasets\\'+str(year)+'\\VIOLATN.CSV'
    print(path)
    df=pd.read_csv(path,encoding = "ISO-8859-1", low_memory=False)
    df['YEAR']=year
    df_final_vio=pd.concat([df_final_vio,df]).astype('Int32')
df_final_vio
```

```python
df_final=pd.merge(df_final,df_final_vio,how='left',on=['STATE','ST_CASE','VEH_NO','YEAR']).astype('Int32')
df_final
```
