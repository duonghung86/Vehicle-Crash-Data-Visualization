# Accident.csv

```python
def union_data(fname):
    df_final=pd.DataFrame()
    for year in range(2010,2019,2):
        path='.\Datasets\\'+str(year)
        name=''
        files=os.listdir(path)
        for file in files:
            if file.find(fname)!=-1:
                name = file
                break
        name
        path += '\\'+name
        print(path)
        df=pd.read_csv(path,encoding = "ISO-8859-1", low_memory=False)
        df['YEAR']=year
        print(df.shape)
        df_final=pd.concat([df_final,df])
    return df_final
df_acc=union_data(select_files[0])
```

```python
na_cols=df_acc.isna().any()[df_acc.isna().any()==True].index
na_cols
```

```python
df_acc.drop(columns=na_cols,inplace=True)
df_acc.shape
```

```python
df_acc.to_csv('accident_10ys.csv',index=False)
```

```python
for fname in sfile:
    df=union_data(fname)
    na_cols=df.isna().any()[df.isna().any()==True].index
    print(na_cols)
    df.drop(columns=na_cols,inplace=True)
    print(df.shape)
    df.to_csv('.\Datasets\\'+fname+'_5ys.csv',index=False)
```
